#Convertir de KM a Millas

```python
primer_p = input('Introduzca la distancia del sol para el primer planeta en KM')
segundo_p = input('Introduzca la distancia desde el sol para el segundo planeta en KM')

primer_p = int(primer_p)
segundo_p = int(segundo_p)

distancia_km = segundo_p - primer_p

print('Primer planeta')
print(primer_p)
print('Segundo planeta')
print(segundo_p)

print('Distancia en km')
print(distancia_km)

distancia_mi = distancia_km * 0.621

print('Distancia en millas')
print(abs(distancia_mi))
```
