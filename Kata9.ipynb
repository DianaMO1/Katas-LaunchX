En lugar de simplemente crear el informe, la función también está calculando el promedio. Mejora la legibilidad extrayendo el cálculo promedio de la función en una nueva función para que el promedio se pueda hacer de forma independiente:

#Función promedio 

Ahora actualiza la función de informes para llamando a la nueva función del promedio:

#Actualiza la función

```python
def generate_report(tanque1, tanque2, tanque3):
    promedio_t = (tanque1 + tanque2 + tanque3) / 3
    return f"""informe:
    Promedio total: {promedio_t}%
    tanque1: {tanque1}%
    tanque2: {tanque2}%
    tanque3: {tanque3}% 
    """
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
print(generate_report(80, 70, 85))
# Función promedio 
def average(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items

# Test the averaging function with a list of integers:
average([80, 85, 81]) 
# Actualiza la función
def generate_report(tanque1, tanque2, tanque3):
    return f"""informe:
    Promedio total: {average([tanque1, tanque2, tanque3])}%
    Tanque1: {tanque1}%
    tanque2: {tanque2}%
    tanque3: {tanque3}% 
    """

# Call the updated function again with different values
print(generate_report(88, 76, 70))
```

```python
def mission_report(pre_launch_time, tiempo_v, planeta, tanque1, tanque2):
    return f"""
    Destino {planeta}
    Tiempo total: {pre_launch_time + tiempo_v} minutes
    Combustible total: {tanque2 + tanque1} gallons
    """

print(mission_report(14, 51, "Luna", 200000, 300000))
# Escribe tu nueva función de reporte considerando lo anterior

def mission_report(planeta, *minutes, **fuel_reservoirs):
    return f"""
    Destino {planeta}
    Tiempo total: {sum(minutes)} minutes
    Combustible total: {sum(fuel_reservoirs.values())}
    """

print(mission_report("Luna", 10, 15, 51, tanque1=300000, tanque2=200000))
# Escribe tu nueva función

def mission_report(planeta, *minutes, **fuel_reservoirs):
    main_report = f"""
    Destino {planeta}
    Tiempo total: {sum(minutes)} minutes
    Combustible total: {sum(fuel_reservoirs.values())}
    \n"""
    for tank_name, gallons in fuel_reservoirs.items():
        main_report += f"{tank_name} {gallons} gallons left\n"
    return main_report

print(mission_report("Luna", 8, 11, 55, tanque1=300000, tanque2=200000))
```
